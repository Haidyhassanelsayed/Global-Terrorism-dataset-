```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 3))
plt.plot(attacks_per_year.index, attacks_per_year.values)
plt.title('no. of attacks per year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x=attacks_per_region.index, y=attacks_per_region.values)
plt.title('NO. of Attacks by Region')
plt.xlabel('Region')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=90)
plt.show()
```

```python
numeric_values = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_values.corr()


plt.figure(figsize=(25, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Different Features')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='attacktype1_txt', y='nkill', data=df)
plt.title('Relationship Between Number of Casualties and Type of Attack')
plt.xlabel('Type of Attack')
plt.ylabel('Number of Casualties')
plt.xticks(rotation=90)
plt.show()
```

```python

kill_summary = df[['nkill', 'nkillus']].sum().reset_index()
kill_summary.columns = ['Category', 'Total Count']

plt.figure(figsize=(5, 3))
sns.barplot(x='Category', y='Total Count', data=kill_summary, palette='viridis')
plt.title('Comparison of Total Killings and US Killings')
plt.xlabel('Category')
plt.ylabel('Total Count')
plt.show()
```
